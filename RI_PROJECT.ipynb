<h1>le modèle LSI (Latent Semantic Indexing)</h1>

```python
from gensim import corpora
from gensim.models import LsiModel
from gensim.parsing.preprocessing import preprocess_string
#from gensim.utils import simple_preprocess
```

```python
# Pretraitement des documents
# Tokenization / convertion en minuscules / élimination de La ponctuation, des chiffres et Les mots vides
processed_docs = [preprocess_string(doc['text']) for doc in documents]

# Creation d'un dictionnaire à partir des documents
# eliminer les repetitions
dictionary = corpora.Dictionary(processed_docs)

# Creation d'une representation sous forme de sacs de mots (bag-of-words) des documents
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
# Afficher les 5 premiers documents dans processed_docs
for i in range(min(5, len(processed_docs))):
    print(f"Document {i+1}: {processed_docs[i]}")
```

```python
# Afficher le contenu du dictionnaire
print("Contenu du dictionnaire:")
for token, idx in dictionary.token2id.items():
    print(f"{idx}: {token}")
```

```python
# Afficher la taille du dictionnaire
print(f"Taille du dictionnaire : {len(dictionary)} mots")
```

```python
# Création du modèle LSI
num_topics = 100  # Nombre de sujets (latent semantic topics)
lsi_model = LsiModel(bow_corpus, num_topics=num_topics, id2word=dictionary)


# l'indice du sujet latent , 'et la représentation du sujet en termes de termes pondérés'
topics = lsi_model.print_topics()
for topic in topics:
    print(topic)

# valeur négative : mot est négativement associé au thème ( les relations opposées)
# valeur positive : mot est positivement associé au thème

# les relations sémantiques entre les termes et le thème latent
```

```python
import numpy as np


# la matrice des vecteurs singuliers à gauche
u_matrix = lsi_model.projection.u

# Matrice diagonale des valeurs singulières
sigma_matrix = np.diag(lsi_model.projection.s)

# la matrice des vecteurs singuliers à droite transposée
vt_matrix = lsi_model.projection.u.T

# Affichage des dimensions des matrices
print("Matrice des vecteurs singuliers à gauche (U):")
print(u_matrix)
print("Taille : ", u_matrix.shape)

print("Matrice diagonale des valeurs singulières (Σ):")
print(sigma_matrix)
print("Taille : ", sigma_matrix.shape)

print("Matrice des vecteurs singuliers à droite transposée (V^T):")
print(vt_matrix)
print("Taille : ", vt_matrix.shape)
```

```python
# les vecteurs de sujets pour chaque document, générés à partir du modèle LSI
# Chaque élément du vecteur indique le degré d'association du document à un sujet
document_vectors = [lsi_model[doc] for doc in bow_corpus]

# Affichage des topics (sujets) latents pour chaque document
for i, doc_vector in enumerate(document_vectors):
    print(f"Document {i+1} - Topics: {doc_vector}")
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

query_text ="what methods -dash exact or approximate -dash are presently available for predicting body pressures at angle of attack."

#crétraitement de la requête
processed_query = preprocess_string(query_text)

#convertir la requête en un vecteur de topics
query_bow = dictionary.doc2bow(processed_query)
query_vector = lsi_model[query_bow]

#convertir le vecteur de la requête en une liste
query_vector_list = [0.0] * num_topics
for i, value in query_vector:
    query_vector_list[i] = value

#convertir la liste en un tableau numpy 2D
query_vector_array = np.array(query_vector_list).reshape(1, -1)

#convertir les vecteurs de topics des documents en un tableau numpy 2D
document_vectors_array = np.zeros((len(document_vectors), num_topics))
for i, doc_vector in enumerate(document_vectors):
    for j, value in doc_vector:
        document_vectors_array[i, j] = value



# Afficher les poids des topics associés à la requête
print("Query Topics:")
for i, value in query_vector:
    print(f"Topic {i}: {value}")


```

```python
#calculer la similarité cosinus entre la requête et chaque document
similarities_to_query = cosine_similarity(query_vector_array, document_vectors_array)

#obtenir les indices triés par ordre décroissant
sorted_indices = np.argsort(similarities_to_query[0])[::-1]

#afficher les similarités triées
print("Similarités triées par ordre décroissant:")
for rank, doc_id in enumerate(sorted_indices):
    similarity = similarities_to_query[0, doc_id]
    print(f"Rang {rank + 1}: Document {doc_id + 1}, Similarité: {similarity}")
```

```python
similarity_threshold = 0.4  #Aadjust the threshold as needed
retrieved_documents = [doc_id for doc_id in sorted_indices if similarities_to_query[0, doc_id] >=similarity_threshold]

#pprint the retrieved documents
print("Retrieved Documents:", retrieved_documents)
print("Retrieved Documents:", len(retrieved_documents))
```

```python
def evaluate_retrieval(relevant_documents, retrieved_documents):
    # Convert relevant and retrieved documents to sets for easier comparison
    relevant_set = set(relevant_documents)
    retrieved_set = set(retrieved_documents)

    # True Positives
    true_positives = len(relevant_set.intersection(retrieved_set))

    # False Positives
    false_positives = len(retrieved_set - relevant_set)

    # False Negatives
    false_negatives = len(relevant_set - retrieved_set)

    # Calculate precision, recall, and F-measure
    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) > 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1_score

#Example usage
#relevant_documents = [184, 29, 31, 12, 51, 102, 13, 14, 15, 57, 378, 859, 185, 30, 37, 52, 142, 195, 875, 56, 66, 95, 462, 497, 858, 876, 879, 880, 486]
relevant_document=[48,122,58,20,196,354,360,197,999,1122,1005,492]
precision, recall, f1_score = evaluate_retrieval(relevant_documents, retrieved_documents)

print(f"Précision: {precision:.2f}")
print(f"Rappel: {recall:.2f}")
print(f"F-mesure: {f1_score:.2f}")
```
